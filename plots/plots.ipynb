```python
import numpy as np
import networkx as nx
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from os import path
```

```python
def plot_graph(index, ax):
    pos = index - 1
    best = [20, 22, 23]
    edges = [11, 12, 13]
    graphs = [str(i) for i in range(1, edges[index-1])]
    colors = sns.color_palette("viridis", as_cmap = True)(np.linspace(0, 1, edges[index-1]))
    for graph in graphs:
        y = []
        for p in range(1, 10):
            x = []
            for seed in range(1, 30):
                fname = "../result/delete_3graphs/graph" + str(index) + "/" + graph + "_p_" + str(p) + "_seed_" + str(seed)
                if path.exists(fname):
                    result = pickle.load(open(fname, "rb"))
                    x.append(result['energy']/-best[index-1])
                else:
                    print(fname)
            if x != []:
                y.append(max(x))
        ax[pos].plot(list(range(1, 10)), y, label = graph, c = colors[int(graph)-1])
    y = []    
    for p in range(1, 10):
        x = []
        for seed in range(1, 30):
            fname = "../result/delete_3graphs/graph" + str(index) + "/0_p_" + str(p) + "_seed_" + str(seed)
            if path.exists(fname):
                result = pickle.load(open(fname, "rb"))
                x.append(result['energy']/-best[index-1])
            else:
                print(fname)
        if x != []:
            y.append(max(x))
    ax[pos].plot(list(range(1, 10)), y, label = 'graph' + str(index), marker='o', linestyle='dashed')
    ax[pos].legend(bbox_to_anchor = (0., 1.1, 1., .07), loc = 'lower left', ncol = 4, mode = "expand", borderaxespad = 0.)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (18, 6))
plot_graph(1, ax)
plot_graph(2, ax)
plot_graph(3, ax)
ax[0].set_ylabel('Approximation Ratio', fontsize = 18)
ax[0].set_xlabel('$p$', fontsize = 18)
ax[1].set_xlabel('$p$', fontsize = 18)
ax[2].set_xlabel('$p$', fontsize = 18)
ax[0].set_title('(a) graph1, $|V| = 10, |E| = 30, C_{max} = 20$', fontsize = 18)
ax[1].set_title('(b) graph2, $|V| = 10, |E| = 33, C_{max} = 22$', fontsize = 18)
ax[2].set_title('(c) graph3, $|V| = 10, |E| = 30, C_{max} = 23$', fontsize = 18)
plt.setp(ax, xticks = np.arange(1, 10, 1), xlim = [1, 9])
fig.tight_layout() 
plt.show()
fig.savefig('delete_graph.pdf', format = 'pdf', bbox_inches = 'tight')
```
